**Classes in Python**

```python
class calculate_sal:
    def __init__(self,base_sal,bonus_rate = .1,symbol='$'):
        self.base_sal = base_sal
        self.bonus_rate = bonus_rate
        self.symbol = symbol
        self.total_sal = base_sal * (1 + bonus_rate)
        self.bonus = self.total_sal - self.base_sal

    def __repr__(self):
        return f'{self.symbol}{self.base_sal:,.0f}'
    
    def show_totalsal(self):
        return f'{self.symbol}{self.total_sal:,.0f}'
    
    def show_bonus(self):
        return f'{self.symbol}{self.bonus:,.0f}'

    
```

```python
emp_1 = calculate_sal(10000,.4,'₹')
print(emp_1)
print(emp_1.show_totalsal())
print(emp_1.bonus_rate)
print(emp_1.show_bonus())
```

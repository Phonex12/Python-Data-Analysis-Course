# Filtering Data in pandas

```python
import pandas as pd
df = pd.read_csv('DataSets\\data_jobs.csv')
```

```python
df.info()
```

**Now when we see the csv file the job_posted_date is a String insted of an actual date variable**

* To convert the string to datetime in pandas we can use the pd.datetime(str) and pass the string inside it

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_date'].dt.month
```

**Now to add a new column in the dataframe we can add it in this manner**

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
df.head()
```

# Sorting Out values in pandas

```python
df.sort_values(by = 'job_posted_date' , inplace=True, ignore_index=True)
df
```

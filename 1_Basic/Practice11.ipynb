# Matplotlib in Python

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('DataSets\\data_jobs.csv')
df.info()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
count_job = df['job_posted_date'].value_counts()
count_job = count_job.sort_index(ascending=True)
count_job
```

**Now That we have transferred all the data into the count_job Series we can now plot it using the matplotlab library**

```python
plt.plot(count_job.index,count_job)
plt.title('Number of Jobs')
```

**Diving deep into the data analysis**

```python
df['month_job'] = df['job_posted_date'].dt.month
```

```python
monthly_count = df['month_job'].value_counts()
monthly_count = monthly_count.sort_index()
monthly_count
```

```python
plt.plot(monthly_count.index,monthly_count)
```

# Bar Graphs in Matplotlib

```python
df.info()
```

```python
job_counts = df['job_title_short'].value_counts()
job_counts = job_counts.sort_values(ascending=True)
job_counts
```

**Now we got the required data into series lets now create a bar graph of it**

To create the bar Graph we can use the `bar(x,y)` to plot values vertailly

```python

plt.bar(job_counts.index,job_counts)
plt.xlabel('Name of Jobs')
plt.ylabel('Number of job postings')
plt.xticks(rotation = 45,ha = 'right')
plt.title('Posting by job title')
```

Hear the problem is that while plotting index vertically it is getting merge between each other so in order to fix it we can use the `barm(x,y)` to plot values horizontally

```python

plt.barh(job_counts.index,job_counts)
plt.title('Job Search as per field')
plt.xlabel('Number of Jobs')
plt.ylabel('Names of Jobs')
plt.show()
```

**But instead of using the matplotlib's plot() which takes a x & y we can use the pandas plot(kind) that only takes one argumnet where we just have to define what kind of plot we need and all the property of title tick or lable will the same as before using the plt class**

```python
job_counts.plot(kind = 'barh')
```

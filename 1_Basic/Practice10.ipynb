# Data Analysis in Pandas

```python
import pandas as pd
df = pd.read_csv('DataSets\\data_jobs.csv')
df.info()
```

```python
df.describe()
```

```python
df.count()
```

```python
df['job_title_short'].unique()
```

```python
df['job_title_short'].value_counts()
```

```python
df['salary_hour_avg'].idxmax()
df['salary_hour_avg'].idxmin()
df['salary_hour_avg'].median()
df['salary_hour_avg'].mode()
```

* Now if in case we want to see what is the minimum salary of the specific column so in this case instead of the using the unique() function we can use the group by attrubute 

```python
df.groupby('job_title_short')['salary_year_avg'].mean()
```

* If we want to group multiple columns we can say add the names of the columns in the list
* Also if we want to perform multiple operation like (mean,median,mode,std) we can also add them in the list

```python
df.groupby('job_title_short')[['salary_year_avg','salary_hour_avg']].agg(['mean','median','count'])
```

```python
df['job_country'].value_counts()
```

**To find whether any of the value is persent in the data we can use the isin() method**

```python
df['job_country'].isin(['India'])
```

**To just know in one word that is it present or not we can use the any() method on it**

```python
df['job_country'].isin(['India']).any()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

df_1 = pd.read_excel("D:\\python Programming\\DataSets\\developer_count.xlsx")
df_2 = pd.read_csv("D:\\python Programming\\DataSets\\data_jobs.csv")
```

```python
df_2['job_posted_date'] = pd.to_datetime(df_2['job_posted_date'])
```

```python
df_2.info()
```

```python
df_2['posted_month'] = pd.to_datetime(df_2['job_posted_date']).dt.strftime('%B')
```

```python
df_2_pivot = df_2.pivot_table(index = 'posted_month' ,columns = 'job_title_short',aggfunc='size')
```

```python
df_2_pivot
```

```python
df_2_pivot.reset_index(inplace=True)
```

```python
df_2_pivot['month_no'] = pd.to_datetime(df_2_pivot['posted_month'], format='%B').dt.month
```

```python
df_2_pivot
```

```python
df_2_pivot.set_index('month_no',inplace=True)
```

```python
df_2_pivot.sort_index(inplace = True)
```

```python
df_2_pivot
```

```python
df_2_pivot.set_index('posted_month',inplace=True)
```

```python
df_2_pivot
```

```python
df_1
```

Cleaning the `df_1` columns

```python
df_1['posted_month'] = df_1['Unnamed: 0'] 
```

```python
df_1
```

```python
df_1.drop(columns='Unnamed: 0',inplace=True)
```

```python
df_1
```

## Merging the two DataSets in pandas using the merge

**Before merging the dataset sorting top 3 skills from the second table**

```python
top3 = df_2['job_title_short'].value_counts().head(3)
top3
```

```python
top3 = top3.index.to_list()
```

```python
df_2_pivot = df_2_pivot[top3]
```

```python
df_2_pivot
```

## Finally merging the data

```python
merged_data = df_2_pivot.merge(df_1,on = 'posted_month',how = 'inner')
```

```python
merged_data
```

**Plotting the data**

```python
merged_data.plot(kind = 'line')
plt.xlabel('Months in 2023')
plt.ylabel('Number of jobs')
plt.legend()
plt.title('Top five jobs in 2023')
plt.show()
```

### **Getting started with Pie Charts**

```python
import pandas as pd
from matplotlib import pyplot as plt
import ast 
```

```python
df = pd.read_csv('D:\\python Programming\\DataSets\\data_jobs.csv')
```

```python
df.info()
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie',startangle = 180,autopct = '%1.1f%%')
plt.ylabel('')
plt.title('Work From Home Status')
```

```python
df['job_title_short'].value_counts().plot(kind = 'pie',autopct = '%1.1f%%')
plt.ylabel('')
```

***Now lets generate a subplot for the three boolean value colunm***

```python
df.info()
```

```python
fig,ax = plt.subplots(1,3)

dict_cols = {
    'job_work_from_home':'Job With Work From Home',
    'job_no_degree_mention':'Job with No Degree',
    'job_health_insurance':'Job with Health Insurance'
}

for i,(cols,title) in enumerate(dict_cols.items()):
    ax[i].pie(df[cols].value_counts(),autopct = '%1.1f%%',startangle = 90,labels = ['False','True'])
    ax[i].set_title(title)

fig.tight_layout()
plt.show()
```

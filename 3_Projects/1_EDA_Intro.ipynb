## **EDA DATA INTRODUCTIONAL ANALYSIS** 

**Importing Data**

```python
import pandas as pd
from matplotlib import pyplot as plt
import ast
import seaborn as sns

df = pd.read_csv('D:\\Python Data Analysis Course\\4_DataSet\\data_jobs.csv')
```

**Performing Normal Cleanups**

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x:ast.literal_eval(x) if pd.notna(x) else x)
```

#### **Filtering Data Analysis Role in United States**

```python
df_DA = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')]
```

```python
top_10_loc = df_DA['job_location'].value_counts().head(10).to_frame()
```

```python
sns.set_theme(style='dark')
sns.barplot(data = top_10_loc,x = 'count',y = 'job_location',hue = 'count',palette='dark:b_r',legend = False)
sns.despine()
plt.title('Counts of Job Locations in United States')
plt.xlabel('Number of Jobs')
plt.ylabel('Name of Cities')
```

**Creating a Pie plots**

```python
pie1 = df['job_work_from_home'].value_counts()
pie2 = df['job_no_degree_mention'].value_counts()
pie3 = df['job_health_insurance'].value_counts()
pie1.plot(kind='pie',autopct = '%1.1f%%')

```

```python
pie2.plot(kind='pie',autopct = '%1.1f%%')
```

```python

pie3.plot(kind='pie',autopct = '%1.1f%%')
```

**Plotting all the Pie Charts together**

```python
fig,ax = plt.subplots(1,3,figsize = (10,5))
dict_cols = {
    'job_work_from_home':'Job With Work From Home',
    'job_no_degree_mention':'Job with No Degree',
    'job_health_insurance':'Job with Health Insurance'
}
for i,(cols,title) in enumerate(dict_cols.items()):
    ax[i].pie(df[cols].value_counts(),autopct = '%1.1f%%',startangle = 90,labels = ['False','True'])
    ax[i].set_title(title)

fig.tight_layout()
plt.show()
```
